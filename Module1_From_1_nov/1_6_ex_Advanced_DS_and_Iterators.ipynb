### Exercise 1: *args

```python
def calculate_average(*args):
    Sum = 0
    for num in args: Sum += num
    return Sum / len(args)

def cool_calculate_average(*args):
    if not args:
        return 0   
    return sum(args) / len(args)


print(calculate_average(1, 2, 3, 4, 5))
print(calculate_average(10, 20))
```

### Exercise 2: **kwargs

```python
def print_profile(**kwargs):
    for key, value in kwargs.items():
        print(f'{key}: {value}')
    
print_profile(name="Alex", age=28, city="New York")
```

### Exercise 3: Combining *args and **kwargs

```python
def process_order(customer_name, *items, **details):
    print(customer_name)
    for item in sorted(items): print(item)
    for key, value in details.items(): print(f'{key}->{value}')

process_order("Bob", "Pizza", "Coke", delivery_time="6 PM", address="123 Main St")
```

### Exercise 4: Generator Function

```python
def countdown(start):
    while start > 1:
        yield start
        start -= 1

for num in countdown(5):
    print(num)

gen = countdown(5)
print(next(gen))
```

## Decorators

### Exercise 6: Simple Decorator

```python
def shout(func):
    def make_upper():
        s = func()
        s = s.upper()
        print('Uppering completed')
        return s
    return make_upper
        
def shout_cool(func):
    def make_upper(*args, **kwargs):   
        s = func(*args, **kwargs)
        s = s.upper()
        print('Uppering completed')
        return s
    return make_upper

@shout_cool
def greet(name):
    return f'Hello, I am a {name}'

print(greet('Alice'))
```

### Exercise 7: Decorator with Arguments

```python
def repeat(n):
    def decorator_itself(func):
        def repeating():
            for _ in range(n): func()
        return repeating
    return decorator_itself

@repeat(3)
def say_hi():
    print('Hi!')

say_hi()
```

### Exercise 10: Directory Operations

```python
from pathlib import Path

dir_path = Path("my_new_folder")

dir_path.mkdir(exist_ok=True)

file_path = dir_path / "temp.txt"

file_path.write_text("This is a temp file inside my_new_folder!")

print("Files in directory:")
for item in dir_path.iterdir():
    print(item.name)
```
