# Lambda Functions

```python
x = lambda x, y, z: x + y + z
print(x(2, 3, 4))
```

```python
(lambda x, y, z: x + y + z)(1, 2, 3)
```

## Sorting with key using lambda function

```python
my_list = [('Alice', 30), ('Bob', 25), ('Charlie', 35)]
sorted_list = sorted(my_list, key = lambda t:t[1])
sorted_list[::-1]
```

# Map

```python
numbers = [1, 2, 3, 4]
squared_numbers = map(lambda x:x**2, numbers)
print(list(squared_numbers))
```

# Filter

```python
numbers = [1, 2, 3, 4, 5, 6]
even_numbers = filter(lambda x: x % 2 == 0, numbers)
print(next(even_numbers))

print(list(even_numbers))
```

```python
from functools import reduce as rc

numbers = [1, 2, 3, 4]
product = rc(lambda x, y: x * y, numbers)
print(product)
```
