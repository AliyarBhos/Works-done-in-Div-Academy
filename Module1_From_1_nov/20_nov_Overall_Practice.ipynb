# Task 1: Lambda + map – Sözlərin uzunluğunu tapmaq

```python
words = ['re', 'hello', 'cat']

word_lengths = list(map(lambda word: (word, len(word)), words))

print(tuple(word_lengths))
```

# Task 2: Filter – 4-ə bölünən ədədləri seçmək

```python
numbers = [1, 4, 8, 9, 12, 15, 16]

div_by_4 = list(filter(lambda x: x % 4 == 0, numbers))

print(div_by_4)
```

# Task 3: Dictionary comprehension – Meyvə sözlükləri

```python
fruits = ['apple', 'banana', 'cherry']

fruit_dict = {fruit[0]: fruit for fruit in fruits}

print(fruit_dict)
```

# Task 1: Raise Exception, yash yoxlayici + Email yoxlayici

```python
enter = input('Enter your age: ')
enter_email = input('Enter your email: ')

try:
    enter = int(enter)
except ValueError:
    raise TypeError('Enter valid age')

if '@' not in enter_email:
    raise ValueError('Enter valid email')

print(f'Your age: {enter}, Your email: {enter_email}')
```
