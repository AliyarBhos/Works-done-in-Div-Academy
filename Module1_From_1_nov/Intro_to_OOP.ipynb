### Introduction to OOP in python 

```python
class Dog:
    ear = 2
    tail = 1
    barking = True

    def __init__(self, name):
        self.name = name

class Cat:
    ear = 2
    tail = 1
    barking =  False

    def __init__(self, name):
        self.name = name

my_dog = Dog('Name1')
ur_dog = Dog('Name2')

print(my_dog is ur_dog)
print(my_dog.name)
```

```python
class Car:
    wheel = 4

    def __init__(self, name):
        self.name = name

my_car = Car('JJhb')
my_car.wheel
my_car.name
```

```python
class Cat:
    ear = 2
    tail = 1
    meow = True
    climb = True

    def do_meow(self):
        if self.color != '-':
            return f'I am {self.name} Meow... I am {self.age} years old, my color is {self.color}'
        return f'I am {self.name} Meow... I am {self.age} years old'
    
    def __init__(self, Name, Age, Color='-'):
        self.name = Name
        self.age = Age
        self.color = Color
    
my_cat = Cat('Cat1', 2, 'Black')
your_cat = Cat('Cat2', 3)

my_cat.do_meow()
```

```python
class Math:
    PI = 3.14
    def area_of_circuit(self, radius):
        return Math.PI * radius ** 2
    
    def __init__(self):
        pass

area = Math()
area.area_of_circuit(4)
```

## Inheritance

```python
class British_Cat(Cat):
    meow = False

    ### Overrided method
    def do_meow(self):
        return 'Meowwww'


British_Cat_1 = British_Cat('British', 12)
British_Cat_1.age
British_Cat_1.meow
British_Cat_1.do_meow()
```
