### find primes in the range [10, 20]

```python
def find_primes_in_range(start, end):
    primes = []
    for num in range(start, end + 1):
        if num < 2:
            continue
        for x in range(2, num):
            if num % x == 0:
                break
        else:
            primes.append(num)
    return primes

find_primes_in_range(10, 20)
```

### calculate the nth fibonacci number

```python
def fibonacci(n):
    first = 0
    second = 1
    for _ in range(0, n - 1):
        first, second = second, first + second
    return first

fibonacci(6)
```

### remove duplicates from the list [1, 2, 2, 3, 4, 4, 5]

```python
def remove_duplicates(lst):
    return list(set(lst))

def remove_duplicates2(lst):
    new_lst = []
    for i in lst:
        if i not in new_lst:
            new_lst.append(i)
    return new_lst

lst = [1, 2, 2, 3, 4, 4, 5]
print(remove_duplicates(lst))
print(remove_duplicates2(lst))
```

### --- find_anagrams("listen", ["enlist", "google", "inlets", "banana"])

```python
def find_anagrams(word, words):
    anagrams = []
    for i in words:
        for j in range(len(word)):
            if i[j] not in word:
                break
        else:
            anagrams.append(i)
    return anagrams

lst = ["enlist", "google", "inlets", "banana"]

print(find_anagrams('listen', lst))
```

### check if the number is a perfect square

```python
from math import sqrt
def is_perfect_square(n):
    return int(sqrt(n))**2 == n

def is_perfect_square2(n):
    if n < 0:
        return False
    x = n
    y = (x + 1) // 2
    while y < x:
        x = y
        y = (x + n // x) // 2
    return x * x == n

is_perfect_square(64)
is_perfect_square2(64)
```

### find the longest substring with unique letters in the string "abcdpabcbb"

```python
def longest_unique_substring(string):
    substring = ''
    for x in string:
        if x not in substring:
            substring += (x)
    return substring

string = 'abcdpabcbb'
print(longest_unique_substring(string))
```

### rotate_list([1, 2, 3, 4, 5], 2) -> [4, 5, 1, 2, 3]

```python
def rotate_list(lst, n):
    return lst[-n:] + lst[:-n]

print(rotate_list([1, 2, 3, 4, 5], 2))
```

### check if the list is sorted [1, 2, 3, 4, 5]

```python
def is_sorted(arr):
    for i in range(1, len(arr)):
        if arr[i - 1] > arr[i]:
            return False
    else:
        return True
    
print(is_sorted([1, 2, 3, 5, 6]))
```

### count frequency of elements in the list [1, 2, 2, 3, 3, 3, 4]

```python
def count_frequency(lst):
    for i in range(len(lst)):
        if i > 0 and lst[i] == lst[i - 1]:
            continue
        
        countL = 0
        for j in lst:
            if lst[i] == j:
                countL += 1
                
        print(f"{lst[i]} occurs {countL} times")

def using_dict_find_freq(lst):
    freq = {}
    
    for x in lst:
        if x in freq:
            freq[x] += 1
        else:
            freq[x] = 1
    
    for key, value in freq.items():
        print(f"{key} occurs {value} times")

count_frequency([1, 2, 2, 3, 3, 3, 4])
print('----------'*2)
using_dict_find_freq([1, 2, 2, 3, 3, 3, 3, 4])
```

### implement bubble sort algorithm for the list

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr


a = [64, 34, 25, 12, 22, 11, 90]  
print(bubble_sort(a))
```

### Max Sum

```python
def max_sum_1(lst):
    Sum = []
    for i in range(len(lst)):
        a = lst[i]
        for j in range(i, len(lst)):
            if lst[j] > a:
                b = lst[j]
                Sum.append(a + b)
    return Sum

def max_sum_2(lst):
    return max(lst) + max(lst[:lst.index(max(lst))] + lst[lst.index(max(lst)) + 1:])

def max_sum_3(lst):
    if len(lst) < 2:
        return None 
    first = second = float('-inf')
    for num in lst:
        if num > first:
            second = first
            first = num
        elif num > second:
            second = num
    return first + second



max_sum_3([1, -2, 0, 3])

```

### Write a function to check if given string is an palindrom

```python
def is_palindrom(string):
    n = len(string)
    for i in range(n//2):
        if string[i] != string[n - i - 1]: return False
    return True

is_palindrom('444')
```

### -----------------------Classes--------------------------

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def introduce(self):
        print(f"Hi, I'm {self.name} and I am {self.age} years old.")

class Calculator:
    def add(self, a, b):
        return a + b
    
    def sub(self, a, b):
        return a - b
    
    def mul(self, a, b):
        return a * b
    
    def div(self, a, b):
        return a / b


class Student:
    def __init__(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade
    
    def display(self):
        print(f"Student: {self.name}, Age: {self.age}, Grade: {self.grade}")


class Cat:
    species = 'mammal'
    def __init__(self, name, age):
        self.name = name
        self.age = age

cat1 = Cat("Cat1", 5)
cat2 = Cat("Cat2", 3)
cat3 = Cat("Cat3", 7)

def oldest_cat(*cats):
    return max(cats, key=lambda c: c.age)

oldest = oldest_cat(cat1, cat2, cat3)
print(f"The oldest cat is {oldest.age} years old.")
```
