# Conditional Statements

```python
num1 = 10
num2 = 5


def Program(num1, num2, operator = input('Enter your operator (+, -, *, /)')):
    if operator == '+':
        return num1 + num2
    elif operator == '-':
        return num1 - num2
    elif operator == '*':
        return num1 * num2
    elif operator == '/':
        return num1 / num2
    else: 
        print('Error!!! Your input is not known operation')
    return None

Program(num1, num2)
```

# Loops
### Factorial Calculation

```python
def factorial(n):
    if n == 0:
        return 1
    elif n == 1:
        return 1
    else:
        return n * factorial(n - 1)
    
def factorial_loop(n):
    fac = 1
    while n >= 1:
        fac *= n
        n -= 1
    return fac
print(factorial(5))
print(factorial_loop(5))
```

### Pattern Printing

```python
for i in range(1, 6):
    print('#'*i)

for i in range(5, 0, -1):
    print('#'*i)
```

```python

for j in range(1, 8, 2):
    print(' ' * int((7 - j) / 2) + '*'*j)

def Pyramid(n):
    for num in range(n):
        for _ in range(n - num - 1):
            print(' ', end='')
        for _ in range(2 * num + 1):
            print('*', end='') 
        print()


Pyramid(4)
```
