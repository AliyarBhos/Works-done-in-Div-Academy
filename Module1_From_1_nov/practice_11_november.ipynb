```python
# match
day = 'frida'

match day:
    case 'monday':
        print('1')
    case 'tuesday':
        print('2')
    case 'wednesday':
        print('3')
    case 'thursday':
        print('4')
    case 'friday':
        print('5')
    case 'saturday':
        print('6')
    case 'sunday':
        print('7')
    case _:
        print('TypeError')

    
```

```python
name_age = [('Ali', 20), ('Shahin', 17), ('Fidan', 18)]

for n, a in name_age:
    print(f'{n} ---> {a}')
```
