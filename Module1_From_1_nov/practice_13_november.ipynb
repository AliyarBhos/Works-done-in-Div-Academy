# Set

```python
# 💥 Set magic! Automatically removes duplicates and keeps things unique 😎
my_set = {100, 20, 100, 50}
print(my_set)

my_set.add('student1')
name = ['student1', 'student2']

print(my_set)
print(my_set.intersection(name))
print(my_set.union(name))
```

# Exercise 1: Dictionary Basics

```python
movie = {
    'title': "Movie1",
    'director': "Director1",
    'year': 1990,
    'genre': 'Genre1'
}

print(movie['title'])
movie['rating'] = 8.5
movie['year'] = 1991
movie.pop('genre')
print(movie)
```

```python

movie = {
    'title': "Movie1",
    'director': "Director1",
    'year': 1990,
    'genre': 'Genre1'
}

print(f"Movie Title: {movie['title']}")  


movie.update({
    'year': 1991,
    'rating': 8.5
})


del movie['genre']

print("Final Movie Dictionary:")
for key, value in movie.items():
    print(f"{key}: {value}")
```

# Exercise 5: Set Operations

```python
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

print(set1.union(set2))
print(set1.intersection(set2))
print(set1.difference(set2))
print(set2.difference(set1))
print(set2.symmetric_difference(set1))
print(set1.union(set2).difference(set1.intersection(set2)))
```

# Exercise 10: Combining Concepts

```python
def find_long_words(list, min_length):
    new_list = []
    for word in list:
        if len(word) > min_length:
            new_list.append(word)
    return new_list

words = ["apple", "banana", "cherry", "date", "elderberry"]
find_long_words(words, 5)
```
