# Personalized Len function

```python
def custom_def(enter, count = 0):
    for _ in enter:
        count += 1
    return count

custom_def('Student')
```

## Global Scope


```python
summ = 100

def Sum(x, y):
    global summ
    summ += x + y
    return summ

print(summ) 
print(Sum(10, 20))
```

### Using global variables in function without global keyword

```python
x = 100
y = 200

def Sum(a, b):
    return x + y

Sum(1, 2)  # Even though we order Sum 1 and 2 it still sums the global variables up
```
