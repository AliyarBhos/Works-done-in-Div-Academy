# Simple guessing game:

```python
import random

rand_a = random.randint(1, 10)
rand_b = random.randint(1, 10)

def ask_user(input_a, input_b, rand_a, rand_b):
    return input_a == rand_a and input_b == rand_b

def give_hint(input_a, input_b, rand_a, rand_b):
    if input_a < rand_a:
        print('Increase your first guess')
    if input_a > rand_a:
        print('Decrease your first guess')
    if input_b < rand_b:
        print('Increase your second guess')
    if input_b > rand_b:
        print('Decrease your second guess')

result = 0
while result < 3:
    input_a = int(input('Enter your guess between 1 and 10(1)'))
    input_b = int(input('Enter your guess between 1 and 10(2)'))

    user = ask_user(input_a, input_b, rand_a, rand_b)
    result += 1

    if user:
        print('You won!')
        break
    elif (result < 3):
        give_hint(input_a, input_b, rand_a, rand_b)
        print('Try again!')
else:
    print('Sorry you lost. You only have 3 chances')
```
