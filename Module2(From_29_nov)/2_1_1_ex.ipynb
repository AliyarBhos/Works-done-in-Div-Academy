### Exercise 1: Creating Arrays

```python
import numpy as np

arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
arr3 = np.zeros(10)
arr4 = np.ones((3, 4))
arr5 = np.arange(0, 9)
arr6 = np.linspace(0, 1, 5)

print(arr1)
print(arr2)
print(arr3)
print(arr4)
print(arr5)
print(arr6)
```

### Exercise 2: Array Attributes

```python
random_arr = np.random.randint(1, 100, size=(3, 3))
print(random_arr)
print(random_arr.shape)
print(random_arr.ndim)
print(random_arr.size)
print(random_arr.dtype)
```

### Exercise 3: Indexing and Slicing

```python
a = np.arange(10, 19)
print(f'a = {a}')
print(a[3])
print(a[2:5])

b = np.arange(1, 17).reshape(4, 4)
print(f'b = {b}')
print(b[2, 3])
print(b[:2, -2:])
print(b[:, 2])
```
