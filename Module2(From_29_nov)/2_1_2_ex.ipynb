### Exercise 1: Basic Arithmetic

```python
import numpy as np

x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

print(x + y)
print(3 * x)
print(x ** 2)
```

### Exercise 2: Universal Functions

```python
z = np.array([1, 4, 9, 16, 25])
sqrt_z = np.sqrt(z)

angles = np.array([0, np.pi/2, np.pi])
sin_angles = np.sin(angles)

print(sqrt_z)
print(sin_angles)
```

### Exercise 3: Broadcasting

```python
matrix = np.arange(1, 10).reshape(3, 3)
vector = np.array([10, 20, 30])
different_sized_vector = np.array([100, 200])

Sum = vector + matrix
print(Sum)

try:
    print(different_sized_vector + matrix)
except Exception as e:
    print(e)
```

### Exercise 4: Aggregate Functions

```python
data = np.random.randint(1, 100, size=(3, 5))
print(data)

print(np.sum(data))
print(np.mean(data))
print(np.max(data))

print(np.sum(data, axis = 0))
print(np.mean(data, axis = 1))
```

### Exercise 5: Reshaping Arrays

```python
arr = np.linspace(0, 11, 12)
arr2 = arr.reshape(3, 4)
arr3 = arr2.reshape(4, 3)
arr4 = arr3.reshape(2, 2, 3)

print(arr)
print('\n')
print(arr2)
print('\n')
print(arr3)
print('\n')
print(arr4)
```

### Exercise 6: Stacking Arrays

```python
m1 = np.array([[1, 2], [3, 4]])
m2 = np.array([[5, 6], [7, 8]])

mh = np.hstack((m1, m2))
mv = np.vstack((m1, m2))

print(mh)
print(mv)
```

### Exercise 7: Boolean Indexing

```python
nums = np.arange(-5, 6, 1)

mask = nums > 0
print(nums[mask])
nums[nums < 0] = 0 
print(nums)
```

### Exercise 8: Dot Product and Matrix Multiplication

```python
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

print(f'Dot product of v1 and v2 = {v1 @ v2}')

mat1 = np.arange(1, 5, 1).reshape(2, 2)
mat2 = np.arange(5, 9, 1).reshape(2, 2)

print(f'Dot product of mat1 and mat2 with @')
print(mat1 @ mat2)
print(f'Dot product of mat1 and mat2 with np.dot')
print(np.dot(mat1, mat2))
```
