### Exercise 1: Create a DataFrame from a Dictionary

```python
import pandas as pd

Dict = {
    'Product': ['Apples', 'Oranges', 'Bananas'],
    'Price': [1.5, 1.2, 1.8],
    'InStock': [True, True, False]
}

Df_Dict = pd.DataFrame(data=Dict)

Df_Dict
```

### Exercise 2: Create a DataFrame from a List of Lists

```python
List = [[1, 'Alice', 100000], [2, 'Bob', 80000], [3, 'Charlie', 120000]]
Index_Col = ['ID', 'Name', 'Salary']

Df = pd.DataFrame(data=List, columns=Index_Col)

Df
```

### Exercise 3: CSV File I/O

```python
Df_Dict.to_csv('products.csv', index=False)
```

```python
df_from_csv = pd.read_csv('products.csv')
df_from_csv
```

### Exercise 5: Create a Multi-Index DataFrame

```python
data = {
    ('Q1', 'Sales'): {('USA', 'East'): 100, ('USA', 'West'): 120, ('EU', 'North'): 80},
    ('Q1', 'Profit'): {('USA', 'East'): 20, ('USA', 'West'): 25, ('EU', 'North'): 15},
    ('Q2', 'Sales'): {('USA', 'East'): 110, ('USA', 'West'): 130, ('EU', 'North'): 90},
    ('Q2', 'Profit'): {('USA', 'East'): 22, ('USA', 'West'): 28, ('EU', 'North'): 18}
}

df_from_data = pd.DataFrame(data=data)

df_from_data
```

### Exercise 6: Accessing Data in a Multi-Index DataFrame

### Exercise 9: Renaming Columns and Setting Index

```python
data = {
    'Country': ['Afghanistan', 'Albania', 'Algeria', 'Andorra', 'Angola'],
    'Code': ['AFG', 'ALB', 'DZA', 'AND', 'AGO'],
    'Continent': ['Asia', 'Europe', 'Africa', 'Europe', 'Africa'],
    'Population': [41128771, 2842321, 44903225, 79034, 35588987]
}
df_countries = pd.DataFrame(data)

x = df_countries.rename({'Code': 'ISO_Code'}, axis='columns')
new_df_countries = x.set_index('Country')
new_df_countries
```

### Exercise 10: Multi-Index and Sorting

```python
y = new_df_countries.reset_index()
y
```

```python
final_indexed = y.set_index(['Country', 'Continent'])
final_indexed
```

```python
Sorted = final_indexed.sort_index()

Sorted
```
