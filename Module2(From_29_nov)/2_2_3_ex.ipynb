### Exercise 7: Arithmetic and Aggregation

```python
grades = pd.DataFrame(
    np.random.randint(50, 100, size=(5, 3)),
    columns=['Math', 'Science', 'English'],
    index=['Alice', 'Bob', 'Charlie', 'David', 'Eva']
)

grades + 5
```

```python
cols = ['Math', 'Science', 'English']
grades['Average Grade for each student'] = grades[cols].mean(axis=1)
grades
```

```python
grades.loc['Average Grade for each subject'] = grades[cols].mean(axis=0)
grades
```

### Exercise 8: Handling Missing Data

```python
df_missing = pd.DataFrame({
    'A': [1, 2, np.nan, 4],
    'B': [5, np.nan, np.nan, 8],
    'C': [10, 11, 12, 13]
})

df_missing.isnull()
```

```python
df_missing.dropna(how='any', axis=0)
```

```python
df_missing.fillna(value=0)
```

```python
df_missing_cols = df_missing.columns
df_missing.fillna(value=df_missing[df_missing_cols].mean(axis=0, skipna=True))
```
