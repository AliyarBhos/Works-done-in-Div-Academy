### Exercise 1: Concatenating DataFrames

```python
import pandas as pd

df1 = pd.DataFrame({'A': ['A0', 'A1'], 'B': ['B0', 'B1']}, index=[0, 1])
df2 = pd.DataFrame({'A': ['A2', 'A3'], 'B': ['B2', 'B3']}, index=[2, 3])

concat_vert = pd.concat([df1, df2], axis=0)
concat_vert
```

```python
concat_horz = pd.concat([df1, df2], axis=1)
concat_horz
```

```python
df3 = pd.DataFrame({'A': ['A4', 'A5'], 'B': ['B4', 'B5']}, index=[0, 1])
concat_df1_df3_vertic = pd.concat([df1, df3], axis=0, ignore_index=True)
concat_df1_df3_vertic
```

### Exercise 2: Merging DataFrames (Inner, Outer, Left, Right)

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2'], 'A': ['A0', 'A1', 'A2']})
right = pd.DataFrame({'key': ['K0', 'K1', 'K3'], 'B': ['B0', 'B1', 'B3']})

inner_merge = pd.merge(left=left, right=right, how='inner', on='key')
inner_merge
```

```python
outer_merge = pd.merge(left, right, how='outer', on='key')
outer_merge
```

```python
left_merge = pd.merge(left, right, how='left', on='key')
left_merge
```

```python
right_merge = pd.merge(left, right, how='right', on='key')
right_merge
```

### Exercise 3: Merging on Different Keys

```python
df_employees = pd.DataFrame({'emp_id': [1, 2, 3, 4], 'name': ['Alice', 'Bob', 'Charlie', 'David']})
df_salaries = pd.DataFrame({'id': [1, 2, 3, 5], 'salary': [50000, 60000, 70000, 80000]})

result = pd.merge(df_employees, df_salaries, left_on='emp_id', right_on='id')

result
```

### Exercise 6: Aggregate, Filter, Transform, Apply

```python
df = pd.DataFrame({
    'Team': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Points': [10, 15, 10, 20, 10, 5],
    'Assists': [5, 7, 5, 10, 5, 2]
})

grouped = df.groupby('Team')[['Points', 'Assists']].sum()
grouped
```

```python
team_stats = (
    df.groupby('Team')
    .agg(
        Points=('Points', 'mean'),
        Assists=('Assists', 'sum')
        )
)

team_stats[team_stats['Points'] < 12.0]
```

```python
grouped['Points_Standardized'] = grouped[['Points']] - team_stats[['Points']]
grouped
```

```python
def doubles(x):
    if x['Assists'].sum() > 10:
        x['Points'] = x['Points'] * 2
    return x

result = grouped.groupby('Team').apply(doubles)
result = result.reset_index(level=0, drop=True)
result
```

### Exercise 7: Creating Pivot Tables

```python
titanic
```

```python
titanic.pivot_table('survived', index='sex', columns='class', aggfunc='mean')
```
