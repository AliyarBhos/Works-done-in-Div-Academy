### Exercise 7: Resampling

```python
dates = pd.date_range('2023-01-01', periods=100, freq='D')
df = pd.DataFrame({'Value': np.random.randn(100)}, index=dates)
```

```python
df
```

```python
df.resample('ME').mean()
```

```python
df.resample('W').sum()
```

### Exercise 8: Shifting and Rolling Windows

```python
dates = pd.date_range('2023-01-01', periods=100, freq='D')
df = pd.DataFrame({'Value': np.random.randn(100)}, index=dates)
```

```python
df['Shifted_Value'] = df['Value'].shift(1)
```

```python
df
```

```python
df['Rolling_Mean'] = df['Value'].rolling(window=7).mean()
```

```python
df
```

```python
df.iloc[:10, :]
```
