```python
import pandas as pd
from pathlib import Path
csv_path = Path(r'C:\Users\konta\Documents\DIV_Academy\Module2(From_29_nov)\data\world_population.csv')
df = pd.read_csv(csv_path)
df.head()
```

```python
# print(s.value_counts(normalize=True))
# print(f'{int(57 / s.value_counts().sum() * 100)} %')


s = df['2022 Population']

def f(pop):
    if pop > 100_000_000:
        return 'Big country'
    else:
        return 'Small country'

df['Size Category'] = s.apply(f)
df[['Country', '2022 Population', 'Size Category']]
```

```python
import numpy as np
import pandas as pd

s_yash = pd.Series([10,20,30], index=['ali', 'vali', 'gulnar'])
s_boy = pd.Series([150,160,170],index=['ali', 'hasan', 'gulnar'])

df1 = pd.DataFrame({
    'yash' : s_yash,
    'boy' : s_boy,
})

df2 = pd.DataFrame(
	np.array(
        [
            [ 10., 150.],
			[ 30., 170.],
			[ np.nan, 160.],
			[ 20.,  np.nan]
        ]
    ), 
    columns=['yash', 'boy'],
    index=['ali', 'vali', 'hesen', 'gulnar']
)
df2

df1.to_excel('output.xlsx', index=False)
```
