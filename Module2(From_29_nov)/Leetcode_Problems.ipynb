```python
import pandas as pd

def account_summary(users: pd.DataFrame, transactions: pd.DataFrame) -> pd.DataFrame:
    return (
        transactions
        .groupby('account', as_index=False)
        .agg(balance=('amount', 'sum'))
        .query('balance > 10000')
        .merge(users, on='account')
        [['name', 'balance']]
    )
```
