```python
import seaborn as sns
import pandas as pd
import numpy as np

df = sns.load_dataset("titanic")
df.head()



# TAPŞIRIQ 1:
# 1. Yalnız qadın (sex == 'female') sərnişinləri seç
# 2. Yaşı (age) 18-dən böyük olanları saxla
# 3. Yalnız aşağıdakı sütunları göstər:
#    - survived
#    - pclass
#    - age
#    - fare

m = (df['sex'] == 'female') & (df['age'] > 18)

result = df.loc[m, ['survived', 'pclass', 'age', 'fare']]

result

```

```python


# TAPŞIRIQ 2:
# 1. Dataset-in ilk 100 sətrini seç
# 2. Yalnız age, fare və survived sütunlarını götür
# 3. fare < 20 olan sətirlərdə bütün dəyərləri NaN et
#    (sətirləri silmə, yalnız maskala)

import numpy as np

df2 = df.loc[:99, ['age', 'fare', 'survived']]

mask = df2['fare'] < 20

df2.loc[mask, :] = np.nan

df2
```

```python

# TAPŞIRIQ 3:
# Yeni "age_group" adlı sütun yarat:
# - age < 18        → "Child"
# - 18 <= age < 60  → "Adult"
# - age >= 60       → "Senior"
# NaN yaşlar üçün dəyər "Unknown" olsun

conditions = [df['age'] < 18, (df['age'] >= 18) & (df['age'] < 60), df['age'] >= 60]

choices = ['Child', 'Adult', 'Senior']

df['age_group'] = np.select(conditions, choices, default='Unknown')

df[['age', 'age_group']]
```

```python
# TAPŞIRIQ 4:
# age_filled adli sütun yarat ve age sütununun boş olanları birbaşa median ilə doldur
df['age_filled'] = df['age'].fillna(df['age'].median())
df
```

```python
# TAPŞIRIQ 5:
# 1. Eyni sex, age və fare olan təkrarlanan sətirləri sil
# 2. Qalan məlumatları:
#    - əvvəl fare azalan
#    - sonra age artan
#    şəkildə sırala
# 3. Yalnız sex, age, fare sütunlarını saxla

result = (df.drop_duplicates(subset=['sex', 'age', 'fare']).sort_values(by=['fare', 'age'], ascending=[False, True])[['sex', 'age', 'fare']])

result
```

```python
# TAPŞIRIQ 6:
# 1. Hər pclass üzrə:
#    - sağ qalmayan (survived == 0) sərnişinlərin sayını tap
# 2. Ən çox itki olan pclass-ı müəyyən et
# pclass üzrə sağ qalmayanların sayı

loss_by_pclass = (df[df['survived'] == 0].groupby('pclass').size())

most_loss_pclass = loss_by_pclass.idxmax()

loss_by_pclass, most_loss_pclass
```
