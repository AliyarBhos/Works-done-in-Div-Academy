```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
rng = np.random.default_rng()
# x = rng.normal(0, 5, size=10)
# y = rng.normal(0, 2, size=10)
data = rng.multivariate_normal(
    mean=[0, 0],
    cov=[[5, 1],
         [1, 2]],
    size=1000
)
```

```python
data.var(axis=0)
```

```python
# variance -> ((arr - arr.mean()) ** 2).mean()
```

```python
df = pd.DataFrame(data, columns=['x', 'y'])
```

```python
df
```

```python
sns.set_theme()
plt.hist(df['x'], alpha=0.7);
plt.hist(df['y'], alpha=0.8);
```

```python
tips = sns.load_dataset('tips')
tips.head()
tips['percentage'] = tips['tip'] / tips['total_bill'] * 100
tips
```

```python
grid = sns.FacetGrid(
    data=tips,
    row='sex',
    col='time',
    margin_titles=True
)
b = np.linspace(0, 50, 20)
grid.map(plt.hist, 'percentage', bins=b)
```

```python
sns.catplot(tips, x='day', y='total_bill', hue='sex')
```

```python
categoric_plot = sns.catplot(tips, x='day', y='total_bill', hue='sex', kind='box')
categoric_plot.set_axis_labels(x_var='Day', y_var='Total Bill')
```
