```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=200, n_features=2, centers=5, cluster_std=0.5, random_state=0)

plt.scatter(X[:, 0], X[:, 1], s=50, c=y, cmap='Dark2_r')
plt.show()
```

```python
blob_center = np.array([
    [ 0.2,  2.3], 
	[-1.5,  2.3], 
	[-2.8,  1.8],
	[-2.8,  2.8], 
	[-2.8,  1.3]
])

blob_std = np.array(
    [0.4, 0.3, 0.1, 0.1, 0.1]
)

X, y = make_blobs(
    n_samples=200, 
    n_features=2, 
    centers=blob_center, 
    cluster_std=blob_std,
    random_state=0,
    )

plt.scatter(X[:, 0], X[:, 1], s=20,  c=y, cmap='Accent')
plt.show()
```

```python
def initialize_centroids(X, K):
    rng = np.random.default_rng(110)
    centroids = rng.choice(X, size=K, replace=True)

    return centroids
```

```python
K = 5

rng = np.random.default_rng(110)
centroids = rng.choice(X, size=K, replace=True)

plt.scatter(X[:, 0], X[:, 1], s=20,  c=y, cmap='gist_rainbow', alpha=0.7)
plt.scatter(centroids[:, 0], centroids[:, 1], color='r', s=200, zorder=3, marker='x')
plt.show()
```
