```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=200, n_features=2, centers=5, cluster_std=0.5, random_state=0)

plt.scatter(X[:, 0], X[:, 1], s=50, c=y, cmap='Dark2_r')
plt.show()
```

```python
blob_center = np.array([
    [ 0.2,  2.3], 
	[-1.5,  2.3], 
	[-2.8,  1.8],
	[-2.8,  2.8], 
	[-2.8,  1.3]
])

blob_std = np.array(
    [0.4, 0.3, 0.1, 0.1, 0.1]
)

X, y = make_blobs(
    n_samples=200, 
    n_features=2, 
    centers=blob_center, 
    cluster_std=blob_std,
    random_state=0,
    )

plt.scatter(X[:, 0], X[:, 1], s=20,  c=y, cmap='Accent')
plt.show()
```

```python
def initialize_centroids(X, K):
    rng = np.random.default_rng(110)
    centroids = rng.choice(X, size=K, replace=True)

    return centroids
```

```python
K = 5

rng = np.random.default_rng(110)
centroids = rng.choice(X, size=K, replace=True)

plt.scatter(X[:, 0], X[:, 1], s=20,  c=y, cmap='gist_rainbow', alpha=0.7)
plt.scatter(centroids[:, 0], centroids[:, 1], color='r', s=200, zorder=3, marker='x')
plt.show()
```

```python
a = np.array([[1, 2, 3, 5, 6]])
b = np.array([[10, 12, 12, 10, 10]])

distance = np.sqrt(np.sum(np.square(a - b)))
```

```python
from sklearn.metrics import pairwise_distances

distance2 = pairwise_distances(a, b)
```

```python
distance, distance2
```

```python
pairwise_distances(X, centroids).round(2).min(axis=1)
```

```python
pairwise_distances(X, centroids).round(2).min(axis=1)
```

```python
y_distance = pairwise_distances(X, centroids).round(2).argmin(axis=1)
y_distance
```

```python
y
```

```python
new_clst_center = X[y_distance == 0].mean(axis=0)
```

```python
new_centroids = []

for x in range(K):
    new_centroids.append(X[y_distance == x].mean(axis=0))

new_centroids = np.array(new_centroids)
```

```python
plt.scatter(X[:, 0], X[:, 1], s=20,  c=y, cmap='gist_rainbow', alpha=0.7)
plt.scatter(new_centroids[:, 0], new_centroids[:, 1], color='r', s=200, zorder=3, marker='x')
plt.show()
```

### KMeans

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=5)

y_pred = km.fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], s=20,  c=y, cmap='gist_rainbow', alpha=0.7)
plt.scatter(y_pred[:, 0], y_pred[:, 1], color='r', s=200, zorder=3, marker='x')
plt.show()
```
