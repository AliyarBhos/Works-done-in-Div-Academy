```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(
    n_samples=200,
    n_features=2,
    centers=5,
    cluster_std=0.5, 
    random_state=10,
)
```

```python
plt.scatter(X[:, 0], X[:, 1], s=10)
plt.show()
```

```python
blob_center = np.array([
    [ 0.2, 2.3],
    [-1.5, 2.3],
    [-2.8, 1.8],
    [-2.8, 2.8],
    [-2.8, 1.3]
])

blob_std = np.array([0.4, 0.3, 0.1, 0.1, 0.1])

X, y = make_blobs(
    n_samples=200,
    n_features=2,
    centers=blob_center,
    cluster_std=blob_std,
    random_state=0
)

plt.scatter(X[:, 0], X[:, 1], s=10)
plt.show()
```

```python
K = 5

rng = np.random.default_rng(140)
centroids = rng.choice(X, size=K, replace=False)

plt.scatter(X[:, 0], X[:, 1], s=10)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=200)
plt.show()
```

```python
centroids
```

```python
rng = np.random.default_rng(140)

def initialize_centroid(X, K, rng):
    centroids = rng.choice(X, size=K, replace=False)

    return centroids
```

```python
initialize_centroid(X, 5, rng)
```

```python
from sklearn.metrics import pairwise_distances, pairwise_distances_argmin

y_pred = pairwise_distances_argmin(X, centroids)
```

```python
y_pred
```

```python
plt.scatter(X[:, 0], X[:, 1], s=10, c=y_pred)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=200)
plt.show()
```

```python
def assign_clusters(X, centroids):
    y_pred = pairwise_distances_argmin(X, centroids)

    return y_pred

assign_clusters(X, centroids)
```

```python
new_centroids = []

for i in range(K):
    new_centroid = X[y_pred == i].mean(axis=0)
    new_centroids.append(new_centroid)

new_centroids = np.array(new_centroids)

plt.scatter(X[:, 0], X[:, 1], s=10, c=y_pred)

plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=200, c='red')
plt.scatter(new_centroids[:, 0], new_centroids[:, 1], marker='x', s=200)
plt.show()
```

```python
def update_centroids(X, y_pred, K):
    new_centroids = []

    for i in range(K):
        new_centroid = X[y_pred == i].mean(axis=0)
        new_centroids.append(new_centroid)
    
    new_centroids = np.array(new_centroids)

    return new_centroids

update_centroids(X, y_pred, 5)
```

```python
max_iter = 4
K = 5
rng = np.random.default_rng(140)
centroids = initialize_centroid(X, K, rng)

fig, axs = plt.subplots(2, 2, figsize=(10, 10))

for i in range(max_iter):
    y_pred = assign_clusters(X, centroids)
    new_centroids = update_centroids(X, y_pred, K)

    axs.flatten()[i].scatter(X[:, 0], X[:, 1], s=10, c=y_pred)
    axs.flatten()[i].scatter(centroids[:, 0], centroids[:, 1], s=200, marker='x', c='r')

    centroids = new_centroids
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=5, random_state=10)
y_pred = km.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], s=10, c=y_pred)
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], marker='x', s=200)
plt.show()
```
