```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder


print("Original Data:")
print(df.head(3))
print("\n")

item = LabelEncoder()
df['Item_Label'] = item.fit_transform(df['Item'])
print("After Label Encoding")
print(df.head(3))
print("\n")
```

```python
ohe = OneHotEncoder(sparse_output=False)
encoded_cols = ohe.fit_transform(df[['Item']])
encoded_df = pd.DataFrame(encoded_cols, columns=ohe.get_feature_names_out(['Item']))


print("After OneHot Encoding")
print(encoded_df.head())
```
