### System: 
### 170theta1 + 25theta2 + theta3 = 7
### 180theta1 + 30theta2 + theta3 = 8
### 200theta1 + 35theta2 + theta3 = 12


```python
X = np.array([
    [170, 25, 1],
    [180, 30, 1],
    [200, 35, 1]
])
y = np.array([7, 8, 12])

theta = np.linalg.inv(X) @ y
theta
```

```python
theta_lina = np.linalg.solve(X, y)
theta_lina
```
