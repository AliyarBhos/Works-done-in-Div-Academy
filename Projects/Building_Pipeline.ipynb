```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline


X = df.drop("charges", axis=1)
y = df["charges"]   

num_cols = ['age', 'bmi', 'children']
cat_cols = ['sex', 'smoker', 'region']

num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

cat_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

processing = ColumnTransformer([
    ('num', num_pipeline, num_cols),
    ('cat', cat_pipeline, cat_cols)
])

X_processed = processing.fit_transform(X)
```

```python
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline

df1["InvoiceDate"] = pd.to_datetime(df1["InvoiceDate"])

df1["InvoiceYear"] = df1["InvoiceDate"].dt.year
df1["InvoiceMonth"] = df1["InvoiceDate"].dt.month
df1["InvoiceDay"] = df1["InvoiceDate"].dt.day
df1["InvoiceHour"] = df1["InvoiceDate"].dt.hour
df1["DayOfWeek"] = df1["InvoiceDate"].dt.dayofweek

df1 = df1.drop(["InvoiceNo", "Description", "InvoiceDate"], axis=1)


X = df1.drop("UnitPrice", axis=1)
y = df1["UnitPrice"]


num_cols = ["Quantity","InvoiceYear","InvoiceMonth","InvoiceDay","InvoiceHour","DayOfWeek"]

cat_cols = ["Country","StockCode"]

X[cat_cols] = X[cat_cols].astype(str)

num_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

cat_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder())
])

processing = ColumnTransformer([
    ("num", num_pipeline, num_cols),
    ("cat", cat_pipeline, cat_cols)
])

X_processed = processing.fit_transform(X)
```
