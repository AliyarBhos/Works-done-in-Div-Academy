```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
ss = StandardScaler()
output_ss = ss.fit_transform(df[['DistanceFromHome']])
mm = MinMaxScaler(feature_range=(-1,1))
output_mm = mm.fit_transform(df[['DistanceFromHome']])

fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(12,8))
axs[0,0].hist(df['DistanceFromHome'], bins=50, color='blue')
axs[0,1].hist(output_ss, bins=50, color='red')
axs[0,2].hist(output_mm, bins=50, color='orange')
fig.show()
```
