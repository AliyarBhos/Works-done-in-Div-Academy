```python
summary = pd.DataFrame({
    "dtype": df.dtypes,
    "missing": df.isnull().sum()
}).reset_index()
summary
```

```python
summary['missing_pct'] = (summary['missing'] / len(df)) * 100
summary = summary.sort_values('missing_pct', ascending=False)
summary
```

```python
drop_cols = summary.loc[summary['missing_pct'] > 70, "index"]
df = df.drop(columns=drop_cols)
```

```python
df.columns
```

```python
df['children'] = df['children'].fillna(0)
```

```python
df['country'] = df['country'].fillna("Unknown")
```

```python
summary = pd.DataFrame({
    "dtype": df.dtypes,
    "missing": df.isnull().sum()
}).reset_index()
summary
```

```python
summary['missing_pct'] = (summary['missing'] / len(df)) * 100
summary = summary.sort_values('missing_pct', ascending=False)
summary
```

```python
df["agent"] = df["agent"].fillna(0)
df["agent"] = df["agent"].astype(int)
```

```python
df['has_agent'] = (df['agent'] != 0).astype(int)
```

```python
df
```

```python
summary = pd.DataFrame({
    "dtype": df.dtypes,
    "missing": df.isnull().sum()
}).reset_index()
summary
```

```python
df['children'] = df['children'].astype(int)
```

```python
df['total_nights'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
df['total_guests'] = df['adults'] + df['children'] + df['babies']
```

```python
df['arrival_date'] = pd.to_datetime(
    df['arrival_date_year'].astype(str) + '-' +
    df['arrival_date_month'] + '-' +
    df['arrival_date_day_of_month'].astype(str),
    errors='coerce'
)
```

```python
df = df.drop(columns=['arrival_date_year', 'arrival_date_month', 'arrival_date_day_of_month'])
```
