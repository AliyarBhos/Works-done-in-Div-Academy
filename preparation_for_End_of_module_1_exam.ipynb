### Questions from ChatGPT

### Task 1 — Square every number

```python
numbers = [1, 2, 3, 4]
square_of_numbers = map(lambda x: x**2, numbers)
print(list(square_of_numbers))
```

### Task 2 — Keep only even numbers

```python
numbers = [3, 8, 11, 4, 10]
filtered_numbers = filter(lambda x:x % 2 == 0, numbers)

print(list(filtered_numbers))
```

### Task 3 — Sort strings by their length

```python
Input = ["apple", "kiwi", "banana", "fig"]

Sorted_Input = sorted(Input, key = lambda x:len(x), reverse=False)
print(Sorted_Input)
```

### Task 4 — Count total characters using reduce

```python
from functools import reduce

Input = ["hello", "world"]
count_of_chars_in_Input = reduce(lambda x, y:len(x) + len(y), Input)

print(count_of_chars_in_Input)
```

### Task 5 — Convert a list of temperatures from °C to °F

```python
C = [0, 10, 20, 30]

F = map(lambda x: int(x * 1.8 + 32), C)

print(list(F))
```

### Task 6 — Filter names that start with a vowel

```python
names = ["Ali", "Mark", "Eren", "Oliver", "Sasha"]
filtered_names = filter(lambda x:x[0] in ['A', 'E', 'I', 'O', 'U'], names)
print(list(filtered_names))
```

### Task 7 — Sort students by grade (descending)

```python
givens = [("Alice", 90), ("Bob", 75), ("Carla", 95)]

sorted_by_grades = sorted(givens, key=lambda x:x[1], reverse=True)

print(sorted_by_grades)
```

### Task 8 — Use reduce to find the max number

```python
Input = [5, 12, 3, 27, 8]

Output = reduce(lambda first, second: first if first > second else second, Input)
print(Output)
```

### Task 9 — Clean messy list of words

```python
messy_list = ["  apple", "banana ", "   ", " kiwi", "pear  "]
cleaned_list = sorted(
    filter(lambda x: x != "", map(lambda x: x.strip(), messy_list))
)

print(cleaned_list)
```

### Task 10 — Word frequency counter (1 line each)

```python
from functools import reduce

words = ["cat", "dog", "cat", "bird", "dog", "dog"]

mapped_words = map(lambda x:(x, 1), words)

from functools import reduce

freq = reduce(
    lambda acc, pair: {**acc, pair[0]: acc.get(pair[0], 0) + pair[1]},
    mapped_words,
    {}
)

print(freq)
```
