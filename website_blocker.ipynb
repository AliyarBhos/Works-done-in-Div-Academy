```python
from time import sleep
from datetime import datetime

host_path = "C:\Windows\System32\drivers\etc\hosts"
redirect = "127.0.0.1"
websites = ["facebook.com", "www.facebook.com"]  # FIXED

while True:
    now = datetime.now()
    start = datetime(now.year, now.month, now.day, 9)
    end   = datetime(now.year, now.month, now.day, 17)

    if start < now < end:
        # Work time → block websites
        with open(host_path, "r+") as file:
            content = file.read()
            for website in websites:
                if website not in content:
                    file.write(f"{redirect}    {website}\n")

    else:
        # Non-work time → unblock websites
        with open(host_path, "r+") as file:
            content = file.readlines()
            file.seek(0)
            for line in content:
                if not any(w in line for w in websites):
                    file.write(line)
            file.truncate()

    sleep(5)
```
